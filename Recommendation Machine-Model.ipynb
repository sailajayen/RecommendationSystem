```python
# Memory based CF - Predicting user based CF (collaborative filtering)
def predict(qty, similarity, type='cust'):
    if type == 'cust':
        mean_cust_qty = qty.mean(axis=1)
        #You use np.newaxis so that mean_cust_qty has same format as qty
        qty_diff = (qty - mean_cust_qty[:, np.newaxis])
        pred = mean_cust_qty[:, np.newaxis] + similarity.dot(qty_diff) / np.array([np.abs(similarity).sum(axis=1)]).T
    elif type == 'prod':
        pred = qty.dot(similarity) / np.array([np.abs(similarity).sum(axis=1)])[:, np.newaxis]
    return pred
```

```python
prod_prediction = predict(train_data_matrix, prod_similarity, type='prod')
cust_prediction = predict(train_data_matrix, cust_similarity, type='cust')
```
