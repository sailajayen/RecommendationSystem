```python
# Impute category dtype values

# Get columns with null values and have less than 100 unique values
df_dropped.columns[(df_dropped.nunique() < 100) & (df_dropped.isnull().mean() > 0)]
```

```python
# Fill null values with mode for categorical columns(Less than or equal to 100 unique values)

cat_cols = df_dropped.columns[(df_dropped.nunique() <= 100) & (df_dropped.isnull().mean() > 0)]

df_dropped[cat_cols]=df_dropped[cat_cols].fillna(df_dropped.mode().iloc[0])
```

```python
# All the columns with unqique values > 100  and having null values
df_dropped.columns[(df_dropped.nunique() > 100) & (df_dropped.isnull().mean() > 0)]
```

```python
# Impute with mode for cat columns > 100 unique values

mod_cols = ['Orders.tracking','Orders.payment_date','Orders.payment_id','Orders.payment_code','Orders.updated_date','Orders.shipped_date',\
        'Products.id','Products.name','Products.display_name','Products.shopping_brand','Products.shopping_mpn','Products.amazon_asin',\
        'Products.google_shopping_id','Products.google_shopping_type','Products.quickbooks_id','Products.qb_edit_sequence',\
        'Products.short_description','Products.long_description','Products.seo_title','Products.seo_url','Products.packaging',\
        'Products.upc','Products.hcpcs', 'Products.shipping_length','Products.shipping_width', 'Products.shipping_height',\
        'Products.family_id','Order_Items.product_id']
df_dropped[mod_cols]=df_dropped[mod_cols].fillna(df_dropped[mod_cols].mode().iloc[0])

# Fill with mean for numeric columns > 100 unique values
int_cols = ['Orders.payment_amount','Order_Items.cost','Products.list_price','Products.price','Products.cost']
df_dropped[int_cols]=df_dropped[int_cols].fillna(df_dropped[int_cols].mean())
```
